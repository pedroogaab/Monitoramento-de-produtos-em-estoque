```python
# Move dir, saved in "root" to > "/content/datasets/SKU-110K/"
import os
from pathlib import Path
import shutil

os.makedirs('/content/datasets/SKU-110K', exist_ok=True)
!mv {path_kaggle} /content/datasets/SKU-110K/

save_drive_dir = Path("/content/drive/MyDrive/visão computacional/dataset")
path_images = Path("/content/datasets/SKU-110K/14/SKU110K_fixed/images")
path_labels = Path("/content/datasets/SKU-110K/14/SKU110K_fixed/labels")
```

```python
import math
import shutil

length_train = len(os.listdir(path_images / "train"))
length_test = len(os.listdir(path_images / "test"))
length_val = len(os.listdir(path_images / "val"))

imgs_train = os.listdir(path_images / "train")
imgs_test = os.listdir(path_images / "test")
imgs_val = os.listdir(path_images / "val")

# Create directories in google drive
for sep_n in range(1, 13):
    path_imgs_train = save_drive_dir / f"lote{sep_n}/images/train"
    path_imgs_test = save_drive_dir / f"lote{sep_n}/images/test"
    path_imgs_val = save_drive_dir / f"lote{sep_n}/images/val"

    path_labels_train = save_drive_dir / f"lote{sep_n}/labels/train"
    path_labels_test = save_drive_dir / f"lote{sep_n}/labels/test"
    path_labels_val = save_drive_dir / f"lote{sep_n}/labels/val"

    os.makedirs(path_imgs_train, exist_ok=True)
    os.makedirs(path_imgs_test, exist_ok=True)
    os.makedirs(path_imgs_val, exist_ok=True)

    os.makedirs(path_labels_train, exist_ok=True)
    os.makedirs(path_labels_test, exist_ok=True)
    os.makedirs(path_labels_val, exist_ok=True)

    # Function to split the files in the directory into 12 parts.
    def split_indices(total, parts, n):
        step = math.ceil(total / parts)
        start = (n - 1) * step
        end = min(n * step, total)
        return start, end

    # TRAIN
    start, end = split_indices(length_train, 12, sep_n)
    for idx_train in range(start, end):
        src_img = path_images / "train" / imgs_train[idx_train]
        dst_img = path_imgs_train / imgs_train[idx_train]
        shutil.copy2(src_img, dst_img)

        format_label = imgs_train[idx_train].split(".")[0] + ".txt"
        src_label = path_labels / "train" / format_label
        dst_label = path_labels_train / format_label
        shutil.copy2(src_label, dst_label)

    # TEST
    start, end = split_indices(length_test, 12, sep_n)
    for idx_test in range(start, end):
        src_img = path_images / "test" / imgs_test[idx_test]
        dst_img = path_imgs_test / imgs_test[idx_test]
        shutil.copy2(src_img, dst_img)

        format_label = imgs_test[idx_test].split(".")[0] + ".txt"
        src_label = path_labels / "test" / format_label
        dst_label = path_labels_test / format_label
        shutil.copy2(src_label, dst_label)

    # VAL
    start, end = split_indices(length_val, 12, sep_n)
    for idx_val in range(start, end):
        src_img = path_images / "val" / imgs_val[idx_val]
        dst_img = path_imgs_val / imgs_val[idx_val]
        shutil.copy2(src_img, dst_img)

        format_label = imgs_val[idx_val].split(".")[0] + ".txt"
        src_label = path_labels / "val" / format_label
        dst_label = path_labels_val / format_label
        shutil.copy2(src_label, dst_label)
```
